# **Analysis betwen the no of students with marks obtained in each subject**

```python
plt.style.use('default')
plt.figure(figsize=(14,10))
plt.hist([data['math score'],data['reading score'],data['writing score']],label=['Math Score','Reading Score','Writing Score'])
plt.legend()
plt.grid(True)
plt.title('Number of Students against Scores',fontsize=20)
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()
```

Large number of students have more marks in reading and writing. This means that it is more difficult to score marks in Maths.

So, number of students decreases as marks of students are increasing.

```python
plt.style.use('bmh')
plt.figure(figsize=(10,10))
plt.hist(data['total score'],label=['Total Score'],color=['#30d28a'],bins=12,rwidth=0.9,alpha=1)
plt.legend()
plt.title('Histogram of total score vs frequency',fontsize=20)
plt.show()
```

This shows that total marks of students has a peak near 200 marks

```python
plt.style.use('fast')

plt.figure(figsize=(15,5))
plt.subplot(1,3,1)
plt.scatter(data['math score'],data['reading score'],color=['red'])
plt.xlabel("Marks in Math")
plt.ylabel("Marks in Reading")

plt.subplot(1,3,2)
plt.scatter(data['writing score'],data['reading score'],color='green')
plt.xlabel("Marks in Writing")
plt.ylabel("Marks in Reading")


plt.subplot(1,3,3)
plt.scatter(data['math score'],data['writing score'])
plt.xlabel("Marks in Math")
plt.ylabel("Marks in Writing")

plt.suptitle('Comparison of marks in one subject with another',fontsize=20)
plt.show()
```

The above graphs show that there is high relation between marks in reading and marks in writing as density of plot is high there.

```python
plt.style.use('seaborn')

plt.figure(figsize=(10,15))
plt.subplot(3,1,1)
sns.distplot(data['math score'],kde_kws={"color": "k", "lw": 2, "label": "KDE"},hist_kws={"histtype": "step", "linewidth": 3,"alpha": 1, "color": "g"},bins=14)
plt.xlabel('Math Score')
plt.ylabel('Density')

plt.subplot(3,1,2)
sns.distplot(data['reading score'],kde_kws={"color": "k", "lw": 2, "label": "KDE"},hist_kws={"histtype": "step", "linewidth": 3,"alpha": 1, "color": "g"},bins=14)
plt.xlabel('Reading Score')
plt.ylabel('Density')

plt.subplot(3,1,3)
sns.distplot(data['writing score'],kde_kws={"color": "k", "lw": 2, "label": "KDE"},hist_kws={"histtype": "step", "linewidth": 3,"alpha": 1, "color": "g"},bins=14)
plt.xlabel('Writing Score')
plt.ylabel('Density')

plt.show()
```

 This shows that peak of average marks is around 60 to 70 for each subject

```python
y = ['Math','Reading','Writing']
width = [data['math score'].mean(),data['reading score'].mean(),data['writing score'].mean()]
plt.style.use('dark_background')

plt.figure(figsize = (12,6))
plt.grid(False)
plt.barh(y = y, width = width,color='yellow')
plt.title('Average Scores Per Subject',fontsize=20)
plt.xlabel('Score',fontsize=14)
plt.ylabel('Subject',fontsize=14)
for i,v in enumerate(width):
    plt.text(v, i, " "+str(round(v,2)), color='white', va='center', fontweight='bold')
plt.show()
```
